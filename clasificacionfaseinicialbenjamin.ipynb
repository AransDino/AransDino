````python
import json
import pandas as pd

def mostrar_clasificacion_con_estilos(ruta_archivo):
    # Cargar el archivo JSON con la codificación correcta
    with open(ruta_archivo, 'r', encoding='ISO-8859-1') as archivo:
        datos = json.load(archivo)

    # Crear un DataFrame de pandas con los datos
    df = pd.DataFrame(datos["teams"])

    # Función para aplicar estilos de color a las filas
    def aplicar_estilo(fila):
        puesto = int(fila["Puesto"][:-1])
        if puesto <= 7:
            if puesto % 2 == 0:
                color = 'background-color: lightgreen'
            else:
                color = 'background-color: palegreen'
        elif 8 <= puesto <= 32:
            if puesto % 2 == 0:
                color = 'background-color: lightyellow'
            else:
                color = 'background-color: lemonchiffon'
        elif 33 <= puesto <= 42:
            if puesto % 2 == 0:
                color = 'background-color: lightgrey'
            else:
                color = 'background-color: gainsboro'
        else:
            color = ''
        return [color] * len(fila)

    # Aplicar los estilos de color
    df_estilizado = df.style.apply(aplicar_estilo, axis=1)

    return df_estilizado

# Asegúrate de usar el nombre correcto del archivo que subiste
ruta_archivo = "ClasificacionBenjamin.json"
clasificacion_estilizada = mostrar_clasificacion_con_estilos(ruta_archivo)
clasificacion_estilizada
````

````python
import json
import pandas as pd

def mostrar_clasificacion_con_estilos(ruta_archivo):
    # Cargar el archivo JSON con la codificación correcta
    with open(ruta_archivo, 'r', encoding='ISO-8859-1') as archivo:
        datos = json.load(archivo)

    # Crear un DataFrame de pandas con los datos
    df = pd.DataFrame(datos["teams"])

    # Función para aplicar estilos de color a las filas
    def aplicar_estilo_filas(fila):
        puesto = int(fila["Puesto"][:-1])
        if puesto <= 7:
            if puesto % 2 == 0:
                color = 'background-color: lightgreen'
            else:
                color = 'background-color: palegreen'
        elif 8 <= puesto <= 32:
            if puesto % 2 == 0:
                color = 'background-color: lightyellow'
            else:
                color = 'background-color: lemonchiffon'
        elif 33 <= puesto <= 42:
            if puesto % 2 == 0:
                color = 'background-color: lightgrey'
            else:
                color = 'background-color: gainsboro'
        else:
            color = ''
        return [color] * len(fila)

    # Función para aplicar estilo de color a la columna de puntos
    def aplicar_estilo_columna_puntos(valor):
        return 'background-color: green; color: white; text-align: center;'

    # Aplicar los estilos de color a las filas
    df_estilizado = df.style.apply(aplicar_estilo_filas, axis=1)

    # Verificar si la columna 'PUNTOS' existe
    if 'PUNTOS' in df.columns:
        # Aplicar estilo a la columna de puntos
        df_estilizado = df_estilizado.applymap(aplicar_estilo_columna_puntos, subset=['PUNTOS'])

    # Centrar los valores de todas las celdas
    df_estilizado = df_estilizado.set_table_styles(
        [{'selector': 'td', 'props': [('text-align', 'center')]}]
    )

    return df_estilizado

# Asegúrate de usar el nombre correcto del archivo que subiste
ruta_archivo = "ClasificacionBenjamin.json"
clasificacion_estilizada = mostrar_clasificacion_con_estilos(ruta_archivo)
clasificacion_estilizada





````

````python
import json
import pandas as pd
from IPython.display import display, HTML

def mostrar_resultados_partidos(ruta_archivo):
    # Cargar el archivo JSON con la codificación correcta
    with open(ruta_archivo, 'r', encoding='ISO-8859-1') as archivo:
        datos = json.load(archivo)

    # Crear un DataFrame de pandas con los datos
    df = pd.DataFrame(datos["teams"])

    # Añadir columnas para logos locales y visitantes (futuras)
    df['logolocal'] = ""  # Reemplazar con las rutas de los logos de los equipos locales
    df['logovisitante'] = ""  # Reemplazar con las rutas de los logos de los equipos visitantes

    # Función para generar HTML con logos
    def generar_html_con_logos(row):
        html = f'''
        <tr>
            <td>{row["FECHA"]}</td>
            <td>{row["HORA"]}</td>
            <td><img src="{row["logolocal"]}" alt="{row["Nombre Equipo Local"]}" style="height:50px;"></td>
            <td>{row["Nombre Equipo Local"]}</td>
            <td>{row["G. Local"]}</td>
            <td>{row["G Visit"]}</td>
            <td>{row["Nombre Equipo Visitante"]}</td>
            <td><img src="{row["logovisitante"]}" alt="{row["Nombre Equipo Visitante"]}" style="height:50px;"></td>
        </tr>
        '''
        return html

    # Generar HTML completo
    html = '''
    <table border="1" style="width:100%; border-collapse:collapse; text-align:center;">
        <thead>
            <tr>
                <th>Fecha</th>
                <th>Hora</th>
                <th>Logo Local</th>
                <th>Equipo Local</th>
                <th>Goles Local</th>
                <th>Goles Visitante</th>
                <th>Equipo Visitante</th>
                <th>Logo Visitante</th>
            </tr>
        </thead>
        <tbody>
    '''

    for _, row in df.iterrows():
        html += generar_html_con_logos(row)

    html += '''
        </tbody>
    </table>
    '''

    display(HTML(html))

# Asegúrate de usar el nombre correcto del archivo que subiste
ruta_archivo = "ResultadosPartidos.json"
mostrar_resultados_partidos(ruta_archivo)
````
